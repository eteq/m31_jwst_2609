```python
import os
from pathlib import Path

import numpy as np

from astroquery.mast import Observations

Observations.login(os.environ['MAST_TOKEN'])
```

```python
obses = Observations.query_criteria(proposal_id=2609, project='JWST')
obses
```

```python
msk = np.array(['NGC6791' in row['target_name'].upper() for row in obses])
n6791_obses = obses[msk]
n6791_obses
```

```python
n6791_prods = Observations.get_product_list(n6791_obses)
n6791_prods
```

```python
n6791_prods.write('NGC6791_dataproducts.ecsv', format='ascii.ecsv', overwrite=True)
```

```python
len(np.unique(n6791_prods['productFilename']))
```

```python
uprods = set()
msk = []
for row in n6791_prods:
    fn = row['productFilename']
    msk.append(fn not in uprods)
    if msk[-1]:
        uprods.add(fn)
msk = np.array(msk)
uprods = n6791_prods[msk]

print(np.sum(uprods['size']) / 1024 / 1024/ 1024, 'GB')
```

```python
prod_results = Observations.download_products(uprods)
```

```python
name_to_local_path = {os.path.split(p['Local Path'])[1]:p['Local Path'] for p in prod_results}
n6791_prods['local_path'] = [name_to_local_path[row['productFilename']] for row in n6791_prods]
```

Note the above will fail if any file names don't match.

```python
for row in n6791_prods:
    product_path = Path(os.path.join('mastDownload', row['obs_collection'], row['obs_id'], row['productFilename']))
    target_path = Path(row['local_path']).resolve()
    if not product_path.exists():
        print('linking', product_path, 'to', target_path)
        product_path.symlink_to(target_path)
```
